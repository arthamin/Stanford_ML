```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import SGDRegressor
```

```python
df = pd.read_csv("dataset_6000.csv")
```

```python
X = df.iloc[:, 3:]
Y = df["price"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, random_state=42, test_size=0.2)
```

```python
x_train = x_train.values
y_train = y_train.values
x_test = x_test.values
y_test = y_test.values
```

```python
scale = StandardScaler()

x_train = scale.fit_transform(x_train)
x_test = scale.fit_transform(x_test)
```

```python
sgdr = SGDRegressor(max_iter = 100000)
```

```python
sgdr.fit(x_train, y_train)

w = sgdr.coef_
b = sgdr.intercept_
```

```python
f_wb = sgdr.predict(x_test)

for i in range(x_test.shape[0]):

    print(f"Prediction: {f_wb[i]:0.4f}, Actual value: {y_test[i]:0.4f}")
```

```python
fig, ax = plt.subplots(1, x_test.shape[1], sharey=True, figsize=(15,3))

for i in range(len(ax)):

    ax[i].scatter(x_test[:, i], f_wb, label='Prediction', color='b')
    ax[i].scatter(x_test[:, i], y_test, label='Target', color='r')

fig.set_label("Targte verus Prediction")
plt.legend()
plt.show()
```
