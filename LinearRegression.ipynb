```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([1.0, 2.0, 5.0])
y_train = np.array([300.0, 400.0, 700.0])

m = x_train.shape[0]

i = 0
x_i = x_train[i]
y_i = y_train[i]
```

```python
plt.scatter(x_train, y_train, marker='x', c='r')
plt.title("Housing price")
plt.xlabel("Size (m^2)")
plt.ylabel("Price (1000$)")
plt.show()
```

<b> Above you can see actual values without the prediction </b>

```python
w = 101
b = 200
```

```python
def compute_model_output(x, w, b):

    m = x.shape[0]

    f_wb = np.zeros(m)

    for i in range(m):
        f_wb[i] = w*x[i] + b
    
    return f_wb
```

```python
tmp_f_wb = compute_model_output(x_train ,w, b)

plt.plot(x_train, tmp_f_wb, color='b', label="Prediction")
plt.scatter(x_train, y_train, color='r', marker='x', label="Actual values")
plt.title("Housing price")
plt.xlabel("Size (m^2)")
plt.ylabel("Price 1000$")
plt.legend()
plt.show()
```

<b> Above you can see a simpliest example of Linear Regression model <i>(untrained)</i> </b>
